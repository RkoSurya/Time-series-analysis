## 1.shifting or Lag:

```python
data.Nile.shift(1) # shifting value from given number
```

```python
cols = ['Shift_1','Shift_2','Shift_3','Shift_4','Shift_5','Shift_6','Shift_7','Shift_8','Shift_9','Shift_10']
for i,j in zip(cols,list(range(1,11))):
    data[i] = data.Nile.shift(j)
```

```python
data
```

#### plot shift:

```python
cols = ['Shift_1','Shift_2','Shift_3','Shift_4','Shift_5','Shift_6','Shift_7','Shift_8','Shift_9','Shift_10']
for i,j in zip(cols,list(range(1,11))):
    data[i] = data.Nile.shift(j).plot()
```

## 2.Difference :
value - nan = nan

```python
diff1 = data['Nile'] - data['Shift_1']
```

```python
diff1
```

```python
diff = ['diff_1','diff_2','diff_3','diff_4','diff_5','diff_6','diff_7','diff_8','diff_9','diff_10']
for i,j in zip(cols,diff):
    data[j] = data['Nile'] - data[i]
```

## Another Methods for finding stationrity:

### 1.AD FULLER TEST:


### This testing used to find stationality or not:-

```python
from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(data['Nile'])
```

### 1.statistic value ,  2.p-value,   %=alpha,   %value= is a critical value

```python
result
```

### compare above (p value < 0.05) = it's a reject  null hypothesis

```python
adfuller(data['diff_1'].dropna())     # reject null hypothesis
```

```python
adfuller(data['Shift_1'].dropna())    # reject null hypothesis
```
