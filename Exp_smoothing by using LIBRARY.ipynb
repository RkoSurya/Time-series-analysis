```python
import pandas as pd
import numpy as np
import pydataset as py
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = py.data('Nile')
```

```python
data = data.reset_index(drop=True)
```

## 1.SINGLE EXP SMOOTHING:

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing 
```

```python
model_1 = SimpleExpSmoothing(data.Nile).fit(smoothing_level=0.4)   #smoothing_level= alpha
```

```python
k=model_1.fittedvalues
```

```python
k
```

## Fittedfcast:

## *An array of both the fitted values and forecast values*

```python
h= model_1.fittedfcast
```

```python
h
```

```python
data['Nile'].plot()
plt.plot(model_1.fittedfcast)
plt.rcParams['figure.figsize'] = [4,4]
plt.legend(['Nile','fitted_fcast'])
```

## 2.DOUBLE EXP SMOOTHING or holts:

```python
from statsmodels.tsa.holtwinters import Holt
```

```python
model_2 = Holt(data.Nile).fit(smoothing_level = 0.5,smoothing_trend=0.4)    #smoothing_trend=BETA
```

```python
k =  model_2.fittedvalues
```

```python
k
```

```python
h = model_2.fittedfcast
```

```python
h
```

```python
data['Nile'].plot()
plt.plot(model_2.fittedfcast)
plt.rcParams['figure.figsize'] = [4,4]
plt.legend(['Nile','fitted_fcast'])
```

## 3.triple exp smoothing or holts winter or Exponential smoothing: 

```python
from statsmodels.tsa.api import ExponentialSmoothing
```

```python
#from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
model_3 =  ExponentialSmoothing(data.Nile).fit(smoothing_level = 0.5,smoothing_trend=0.4,smoothing_seasonal=0.5)
```

```python
k = model_3.fittedvalues    # MODEL PREDICTED VALUES
```

```python
k
```

```python
h = model_3.fittedfcast
```

```python
len(h)
```

```python
data['Nile'].plot()
plt.plot(model_3.fittedfcast)
plt.rcParams['figure.figsize'] = [10,9]
plt.legend(['Nile','fitted_fcast'])
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(data.Nile, model_3.fittedfcast)
```
